```python
# Function to perform arithmetic operations
def calculate(n1, n2, op):
    if op == '+':
        return n1 + n2
    elif op == '-':
        return n1 - n2
    elif op == '*':
        return n1 * n2
    elif op == '/':
        if n2 == 0:
            return "Error: Division by zero"
        return n1 / n2
    elif op == 'exp':
        return n1 ** n2
    else:
        return "Invalid operator"

# Get user input
try:
    print("basic calculator")
    num1 = float(input("Enter first number: "))
    operator = input("Enter operator (+, -, *, /,exp): ")
    num2 = float(input("Enter second number: "))
    # Perform and print result
    result = calculate(num1, num2, operator)
    print(f"Result: {result}")
except ValueError:
    print("Invalid input! Please enter numeric values.")
```

```python
def count_numeric(data_list):
  count = sum(1 for item in data_list if isinstance(item, (int, float)) and not isinstance(item, bool))
    return count

# Example Usage:
my_list = [1, 'apple', 3.14, True, 40, 'banana', 5, False]
numeric_count = count_numeric(my_list)

print(f"Original list: {my_list}")
print(f"Number of numeric values (int/float): {numeric_count}")
```

```python
mixed_dict = {'a': 26, 'b': 'hello', 'c': 24, 'd': 5.5, 'e': 'world'}
total_sum = 0

for value in mixed_dict.values():
    # Check if the value is a number (integer or float)
    if isinstance(value, (int, float)):
        total_sum += value

print(total_sum)
```

```python
def find_duplicates_with_loop(input_list):
    unique_items = []
    duplicates = []
    for item in input_list:
        if item in unique_items:
            if item not in duplicates: # Ensure each duplicate is added only once
                duplicates.append(item)
        else:
            unique_items.append(item)
    return duplicates

# Example Usage:
my_list = [1, 65, 13, 44, 5, 2, 65, 13, 1]
print(f"Original list: {my_list}")
print(f"Duplicate elements: {find_duplicates_with_loop(my_list)}")



```
