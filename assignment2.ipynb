```python
#program to sum all numeric items in list
def sum(items):
    total = 0
    for item in items:
        try:
            # Attempt to convert the item to a float (covers ints and floats)
            total += float(item)
        except (ValueError, TypeError):
            # If conversion fails (e.g., a non-numeric string), ignore the item
            pass
    return total

mixed_list = [25, 4, "three", 984, "five", 16]
result = sum(mixed_list)
print(result)

```

```python
#program to get largest mum from numeric list
numbers = [94, 224, 251, 43, 132]
largest = numbers[0]
for n in numbers:
    if n > largest:
        largest = n
print(largest)
```

```python
#program to take list print all mum less than 5

numbersList = [int(x) for x in input("Enter numbers separated by space: ").split()]
print("The list of integers is:", numbersList)
print("The integers less than 5 in the are:")
for n in numbersList:
    if n < 5:
        print(n)
```
